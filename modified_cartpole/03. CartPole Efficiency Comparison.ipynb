# CartPole Efficiency Comparison

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
model_learning_actionshaping = pd.read_csv("logs/ActionShaping/progress.csv")
model_learning_rewardshaping = pd.read_csv("logs/RewardShaping/progress.csv")
```

```python
rewards_actionshaping = np.loadtxt(f"results/ActionShaping/rewards_actionshaping.txt")
steps_actionshaping = np.loadtxt(f"results/ActionShaping/steps_actionshaping.txt")
actions_actionshaping = np.loadtxt(f"results/ActionShaping/actions_actionshaping.txt")
rewards_rewardshaping = np.loadtxt(f"results/RewardShaping/rewards_rewardshaping.txt")
steps_rewardshaping = np.loadtxt(f"results/RewardShaping/steps_rewardshaping.txt")
actions_rewardshaping = np.loadtxt(f"results/RewardShaping/actions_rewardshaping.txt")
```

# ActionShaping

```python
model_learning_actionshaping["rollout/ep_rew_mean"].plot()
plt.title("ActionShaping Learning Progress")
plt.xlabel("Model Number")
plt.ylabel("Average Reward")
```

```python
print(f"Reward mean: {rewards_actionshaping.mean()}")
print(f"Reward std : {rewards_actionshaping.std()}")
```

```python
print(f"Step mean: {steps_actionshaping.mean()}")
print(f"Step std : {steps_actionshaping.std()}")
```

```python
print(f"Number of left pull : {list(actions_actionshaping).count(0)}")
print(f"Number of right pull: {list(actions_actionshaping).count(1)}")
print(f"Number of no pull   : {list(actions_actionshaping).count(2)}")
```

```python
print(f"Percentage of no pull: {list(actions_actionshaping).count(2)/len(actions_actionshaping) * 100}%")
```
